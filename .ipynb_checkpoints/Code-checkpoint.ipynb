```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('data.csv')
data.head(10)
```

```python
country = data.Country[ ~data.Country.isna() ].values

country_checked = data[data.Status == 'Checked Out'].groupby('Country',as_index=False).size()
country_checked = pd.DataFrame(country_checked)
country_checked_new = country_checked[country_checked.values >= 100]

plt.figure()
plt.title('Countries that have Checked out Successfully')

plt.bar( country_checked_new.index, country_checked_new[0],color='navy' )

plt.ylabel('Number of checked outs')
plt.xticks(rotation = 75)
plt.show()
country_checked = country_checked.rename(columns={0:'Checked Out'})
country_checked
```

```python
country_cancelled = data[data.Status == 'Cancelled'].groupby('Country',as_index=False).size()
country_cancelled = pd.DataFrame(country_cancelled)
country_cancelled_new = country_cancelled[country_cancelled.values >= 90]

plt.figure()
plt.title('Countries that have Cancelled their reservations')

plt.bar( country_cancelled_new.index, country_cancelled_new[0],color='pink' )

plt.ylabel('Number of cancellations')
plt.xticks(rotation = 75)
plt.show()
country_cancelled = country_cancelled.rename(columns={0:'Cancelled'})
country_cancelled
```

```python
result = pd.concat([country_cancelled, country_checked],axis=1)
# result
ressult = result.dropna(subset = ["Cancelled"], inplace=True)
# result = result.replace(np.nan,0)
result['new'] = result['Checked Out']/(result['Checked Out'] + result['Cancelled'])
result

result = result[result['new'].values >= 0.97]

plt.figure()
plt.title('Countries that have been most helpful')

plt.bar( result.index, result['new'],color='pink' )

plt.ylabel('Success rate')
plt.xticks(rotation = 75)
plt.show()
```

```python
# Prepare Data
to_plot = data.corr()
n_columns = len(to_plot)

# Create Figure
plt.figure( figsize=(2,2), dpi = 150 )

# Add Heatmap
plt.imshow( to_plot.values, cmap = 'RdBu' )

# Set Aesthetics
plt.xticks( list(range(n_columns)), to_plot.columns, rotation = 45 )
plt.yticks( list(range(n_columns)), to_plot.columns, rotation = 45 )
plt.xlim(-0.6, n_columns-0.4 )
plt.ylim(-0.6, n_columns-0.4)
plt.clim(-1,1)
plt.colorbar()
plt.show()
```

```python
data["check_in_date"]  = pd.to_datetime(data["check_in_date"])
data["check_out_date"]  = pd.to_datetime(data["check_out_date"])
data["booking_date"]  = pd.to_datetime(data["booking_date"])

print(data.dtypes)
data.head(10)
```

```python
plt.figure( figsize=(16,6), dpi = 300 )
stay_duration = data[data.Status == 'Checked Out'].groupby('check_in_date',as_index=False).sum()
stay_duration
```

```python
stay_duration.set_index('check_in_date')['stay_duration'].plot(color='sienna');
```

```python
rooms_types1 = data[data.Status == 'Checked Out'].groupby('room_type_code',as_index=False).size()
rooms_types2 = data[data.Status == 'Checked Out'].groupby('room_type_code',as_index=False).sum()
rooms_types1 = pd.DataFrame(rooms_types1)
rooms_types1 = rooms_types1.rename(columns={0:'Sum'})
rooms_types2 = rooms_types2.set_index('room_type_code')
# rooms_types2
result = pd.concat([rooms_types1, rooms_types2],axis=1)
result
```

```python
plt.figure()
plt.title('Number of reservations of every room type')

plt.bar( result.index, result['Sum'],color='sienna' )

plt.ylabel('Number of checked outs')
plt.xticks(rotation = 75)
plt.show()

# 
result['new'] = result['price_pernight']/(result['Sum'])
result

plt.figure()
plt.title('Profit rate')

plt.bar( result.index, result['new'],color='navy' )

plt.ylabel('$/S')
plt.xticks(rotation = 75)
plt.show()
```
